```
for im in os.listdir('/content/inputs'):
  print("Searching for "+ im.split('.')[0])
  im=face_recognition.load_image_file('/content/inputs/'+im)
  my_encodings = face_recognition.face_encodings(im)[0]
  if(not os.path.exists('saved_images')):
    os.makedirs('saved_images')
  images_to_save=[]
  dir='/content/images/'
  count=0
  list_images=os.listdir(dir)
  for i in range(len(list_images)):
    img=list_images[i]
    image=face_recognition.load_image_file(dir+img)
    list_of_face_encodings = face_recognition.face_encodings(image)
    count+=1
    print(count)
    results = face_recognition.compare_faces(list_of_face_encodings, my_encodings)
    if(True in results):
      print("FOUND -- "+dir+img)
      images_to_save.append(dir+img)
      shutil.copy(dir+img,'/content/saved_images/')
      

  print(len(images_to_save))
  
  
```
